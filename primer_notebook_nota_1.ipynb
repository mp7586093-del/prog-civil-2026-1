# Problema de Ingeniería (Nivel Básico)

Un ingeniero agrícola quiere calcular **el área de un terreno rectangular** para estimar la cantidad de semillas necesarias.

La fórmula es:

[
Área = Largo \times Ancho
]

---

# 1️⃣ Definir **Entradas – Proceso – Salidas**

### Entradas (datos que el usuario proporciona)

* Largo del terreno (m)
* Ancho del terreno (m)

### Proceso (operación matemática)

* Multiplicar el largo por el ancho

### Salidas (resultado del algoritmo)

* Área del terreno en **metros cuadrados (m²)**

---

# 2️⃣ Algoritmo (explicado paso a paso)

1. Pedir al usuario el **largo del terreno**.
2. Pedir al usuario el **ancho del terreno**.
3. Multiplicar ambos valores.
4. Mostrar el área calculada.

---

# 3️⃣ Pseudocódigo

```
INICIO

    LEER largo
    LEER ancho

    area = largo * ancho

    MOSTRAR area

FIN
```

---

# 4️⃣ Código en Python

```python
# ---------------------------------------
# Cálculo del área de un terreno
# Ejemplo básico de variables y operadores
# ---------------------------------------

# Entrada de datos
largo = float(input("Ingrese el largo del terreno (m): "))
ancho = float(input("Ingrese el ancho del terreno (m): "))

# Proceso
area = largo * ancho

# Salida
print("El área del terreno es:", area, "m²")
```

### Un ejemplo aplicado a la Ingeniería

Imagina que necesitas calcular el **volumen de agua** necesario para regar un lote agrícola o para el llenado de un tanque en una obra civil. En lugar de darte el código resuelto, analicemos la lógica:

1. Necesitamos conocer el **área** del terreno ($m^2$).
2. Necesitamos conocer la **lámina de riego** o profundidad deseada ($mm$).
3. El volumen será el producto de ambos (ajustando las unidades).

Aquí tienes un ejemplo de cómo se vería esa lógica estructurada con **buenas prácticas** (variables descriptivas y funciones):

````python
def calcular_volumen_riego(area_m2, lamina_mm):
    """
    Calcula el volumen de agua necesario en metros cúbicos.
    Convierte la lámina de mm a metros (1 mm = 0.001 m).
    """
    lamina_m = lamina_mm / 1000
    volumen_m3 = area_m2 * lamina_m
    return volumen_m3

# Datos de entrada (Imagina que estos vienen de un sensor o un formulario)
area_lote = 500.0  # metros cuadrados
profundidad_deseada = 15.0  # milímetros

# Llamada a la función
resultado = calcular_volumen_riego(area_lote, profundidad_deseada)

print(f"El volumen de agua requerido es: {resultado} m3")
````
